```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds = pd.read_excel('./Анализ мотодов машинного обучения.xlsx')
ds
```

```python
plt.plot(ds['Настроенные продажи'], ds['Unnamed: 0'])
plt.grid()
plt.show()
```

```python
plt.plot(ds['Ненастроенные продажи'], ds['Unnamed: 0'])
plt.grid()
plt.show()
```

```python
class Method:
    def __init__(self, yes, no, saleYes, saleNo):
        self.yes = yes
        self.no = no
        self.saleYes = saleYes
        self.saleNo = saleNo
```

```python
claster = Method(ds['Настроенные запуски'][0], ds['Ненастроенные запуски'][0], ds['Настроенные продажи'][0],ds['Ненастроенные продажи'][0] )
recomend = Method(ds['Настроенные запуски'][1], ds['Ненастроенные запуски'][1], ds['Настроенные продажи'][1],ds['Ненастроенные продажи'][1] )
look = Method(ds['Настроенные запуски'][2], ds['Ненастроенные запуски'][2], ds['Настроенные продажи'][2],ds['Ненастроенные продажи'][2] )
behavior = Method(ds['Настроенные запуски'][3], ds['Ненастроенные запуски'][3], ds['Настроенные продажи'][3],ds['Ненастроенные продажи'][3] )
generic = Method(ds['Настроенные запуски'][4], ds['Ненастроенные запуски'][4], ds['Настроенные продажи'][4],ds['Ненастроенные продажи'][4] )

```

```python
print('Процент успешных продаж с использованием Кластерного анализа:')
```

```python
def calc(a):
    use = (a.saleYes/a.yes)*100
    nope = (a.saleNo/a.no)*100
    ans = [use, nope]
    return ans
```

```python
print('Процент успешных продаж с использованием Кластерного анализа:', calc(claster)[0], '%')
print('Процент успешных продаж без использования Кластерного анализа:', calc(claster)[1], '%')
```

```python
print('Процент успешных продаж с использованием Рекомендательных систем:', calc(recomend)[0], '%')
print('Процент успешных продаж без использования Рекомендательных систем:', calc(recomend)[1], '%')
```

```python
print('Процент успешных продаж с использованием настроек аудитории:', calc(look)[0], '%')
print('Процент успешных продаж без использования настроек аудитории:', calc(look)[1], '%')
```

```python
print('Процент успешных продаж с использованием Прогнозирование поведения пользователей:', calc(behavior)[0], '%')
print('Процент успешных продаж без использования Прогнозирование поведения пользователей:', calc(behavior)[1], '%')
```

```python
print('Процент успешных продаж с использованием Генеративных алгоритмов:', calc(generic)[0], '%')
print('Процент успешных продаж без использования Генеративных алгоритмов:', calc(generic)[1], '%')
```
